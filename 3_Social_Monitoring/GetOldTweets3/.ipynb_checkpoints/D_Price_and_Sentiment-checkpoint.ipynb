```python
#Concat df's into one mega df
frames_2018 = [df_2018_1, df_2018_2, df_2018_3, df_2018_4, df_2018_5, df_2018_6, df_2018_7, df_2018_8, df_2018_9, df_2018_10, df_2018_11, df_2018_12]
frames_2017 = [df_2017_1, df_2017_2, df_2017_3, df_2017_4, df_2017_5, df_2017_6, df_2017_7, df_2017_8, df_2017_9, df_2017_10, df_2017_11, df_2017_12]

df_2018 = pd.concat(frames_2018,axis=0)
df_2017 = pd.concat(frames_2017,axis=0)
df=pd.concat([df_2017, df_2018], axis=0)
df=df.reset_index()
```

```python
df.describe()
```

```python
# Index dates, so that Jan 1sts is '1', ..., Dec 31st is '365', etc. 
def Date_to_Index(year,month,day):
    ref_date = date(2017, 1, 1)
    actual_date = date(year, month, day)
    delta = actual_date - ref_date
    return delta.days
Date_to_Index(2018,1,2)
```

```python
# Get the set of dates that we have the tweet data for
Price_v_DateID_2017_y = df_2017['Price']
test_df=df
test_df=test_df.reset_index()
Set_of_Dates=[[test_df['Year'][i],test_df['Month'][i],test_df['Day'][i]] for i in range(0, test_df['Day'].shape[0])]
# Return the date ID for the Set_of_Dates
Set_of_Dates_ID=[Date_to_Index(Set_of_Dates[i][0], Set_of_Dates[i][1], Set_of_Dates[i][2]) for i in range(0,test_df.shape[0])]
```

```python
#Plot the Price data vs the date ID
xx=Set_of_Dates_ID
yy=test_df['Price']
plt.title('Price Sentiment Analysis') 
plt.xlabel('Date ID (Rel 2017)') 
plt.ylabel('Price') 
plt.scatter(xx,yy)
```

```python
#Plot the sentiment data vs the date ID - NOT NORMALISED
fig = plt.figure()
ax1 = fig.add_subplot(111)
yy_pos  = test_df['Pos']
yy_neg  = test_df['Neg']
yy_tot  = test_df['Total']
ax1.scatter(xx,yy_pos, label='Pos' )
ax1.scatter(xx,yy_neg, label='Neg')
ax1.scatter(xx,yy_tot, label='Total')
plt.legend(loc='upper left');
```

```python
#Plot the sentiment data vs the date ID - NORMALISED
fig = plt.figure()
ax1 = fig.add_subplot(111)
yy_pos  = test_df['Pos']
yy_neg  = test_df['Neg']
yy_tot  = test_df['Total']
senti_nrm=yy_tot
ax1.scatter(xx,yy_pos/senti_nrm, label='Pos' )
ax1.scatter(xx,yy_neg/senti_nrm, label='Neg')
ax1.scatter(xx,yy_tot/senti_nrm, label='Total')
plt.legend(loc='upper left');
```

```python
#Plot the Price and Sentiment vs the date ID - Does the Normalisation make sense?
fig = plt.figure()
ax1 = fig.add_subplot(111)
yy_price=test_df['Price']
yy_pos = test_df['Pos']
yy_neg = test_df['Neg']
yy_tot = test_df['Total']
senti_nrm=yy_tot/yy_price[0]

ax1.scatter(xx,yy_pos/senti_nrm, label='Pos' )
ax1.scatter(xx,yy_neg/senti_nrm, label='Neg')
ax1.scatter(xx,yy_tot/senti_nrm, label='Total')
ax1.scatter(xx,yy_price, label='Price')

plt.legend(loc='upper left');
```
